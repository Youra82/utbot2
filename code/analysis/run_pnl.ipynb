```python
import sys
import os
import json
from pathlib import Path
notebook_dir = os.getcwd()
sys.path.append(os.path.join(notebook_dir, '..'))
from utilities.tax_endpoint_analysis import RecordsManager
```

```python
# --- SETUP ---
# Passe den Pfad zu deiner secret.json an, falls nötig.
key_path = Path(os.getcwd()).parents[1] / 'secret.json'
# Gib hier den Namen des API-Schlüssels an, der für die Analyse verwendet werden soll.
key_name = 'envelope_tax_endpoint'

# Gib das Startdatum an, ab dem die Trades analysiert werden sollen.
portefolio_start_date = "2024-01-01"

# Lade die API-Zugangsdaten
with open(key_path, "r") as f:
    api_setup = json.load(f)[key_name]
    
# Initialisiere den RecordsManager, der die Daten von der Börse holt.
# Optional: Gib einen Dateinamen an, um die Rohdaten als CSV zu speichern (z.B. filename='my_trades')
records_manager = RecordsManager(api_setup, portefolio_start_date, filename='live_trades_utbot2')
```
